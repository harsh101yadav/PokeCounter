```python
import json
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\shubh\Projects\PokeCounter\data.csv")
data.head()
```

```python
data = data.fillna(1)
data.head()
```

```python
data['power1'] = data['power'].apply(lambda x: x.split(" ")[0])
data['power2'] = data['power'].apply(lambda x: x.split(" ")[1])
```

```python
data.head()
```

```python
data = data.drop(['power','PKMN'], axis=1)
data.head()
```

```python
data_dict = data.to_dict(orient="records")
data_dict[0]
```

```python
formatted_data = {}
for i in data_dict:
    p1, p2 = i.pop('power1'), i.pop('power2')
    p1 = p1 if p1 != "—" else "None"
    p2 = p2 if p2 != "—" else "None"

    if p1 not in formatted_data:
        formatted_data[p1] = {p2: {}}
    formatted_data[p1][p2] = i
```

```python
with open("cleaned_data.json", "w") as f:
    json.dump(formatted_data, f, indent= 4)
```
